```python
#how has the average time taken to report changed over time? 
#(motivation, offence type, locations)

def report_time_change(colName,atLeastInstancesForLabels=0):
    '''
        
    '''
    
    #there is an anomaly/error on row 1109, possibly a clerical error
    #we remove this row for this analyis
    hate_crimes_temp = pd.concat([hate_crimes.loc[:1108,:],hate_crimes.loc[1110:,:]])
    
    #select labels with at least X instances
    labels_toconsider = hate_crimes_temp[colName].value_counts()[hate_crimes_temp[colName].value_counts()>=atLeastInstancesForLabels]

    #create a df consisting of years as indices and labels as columns. instances as values
    df_columns = list(labels_toconsider.index)
    df_index = range(2019,2024)
    df = pd.DataFrame(columns=df_columns,index=df_index)
    for label in df_columns:
        for year in df_index:
                df[label].loc[year]=round(hate_crimes_temp.hours_to_report[(hate_crimes_temp.year==year) & (hate_crimes_temp[colName]==label)].mean(),2)
            
    #create the final row that lists change in percentage between 2023 and 2018
    list_of_percentages = []
    list_of_averages = []
    
    for label in df.columns:
        try:
            list_of_percentages.append(round((df.loc[2023,label]-df.loc[2019,label])*100/df.loc[2019,label],2))
            list_of_averages.append(df[label].mean())
        except:
            list_of_percentages.append(0)
            list_of_averages.append(df[label].mean())

    df.loc['Percentage Change']=list_of_percentages
    df.loc['Average'] = list_of_averages
    
    return df
```
